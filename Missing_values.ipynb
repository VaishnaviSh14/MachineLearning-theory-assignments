Missing values
They occur in a dataset,when some of the information is not stored for a variable.
1.Missing completely at random
2.Missing at random
3.Missing data not at random

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#check for missing values
df.isnull().sum()
```

```python
df.shape
```

```python
#Row wise deletion - but it leads to loss of large amount of data
df.dropna().shape
```

```python
#Column wise deletion
df.dropna(axis = 1)
```

```python
#Next technique is imputation missing values
#1.Mean value imputation - replacing null values in the age with the mean value.It works well when we have a normall distributed data

df['age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean','age']]
```

```python
#2.Median value imputation - If we have right skewed or left skewed , then we will also be having outliers in that case we use this
df['age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['age_median','age']]
```

```python
#3.Mode value imputation - used with categorical values
#taking categorical data - embarked from the dataset

df['embarked_mode']=df['embarked'].fillna(df['embarked'].mode())
```

```python
df[['embarked_mode','embarked']]
```
