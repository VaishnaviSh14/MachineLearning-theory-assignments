Feature scaling is the process of normalizing the range of features in a dataset.
Techniques - 
1. Standardization
2. Normalization [MIN MAX SCALER]
3. Unit Vector

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
#standardization using numpy
import numpy as np
```

```python
mean = np.mean(df['total_bill'])
```

```python
std = np.std(df['total_bill'])
print(mean,std)
```

```python
normalized_data = []
for i in list(df['total_bill']):
    z_score = (i-mean)/std
    normalized_data.append(z_score)
```

```python
normalized_data
```

```python
sns.histplot(normalized_data,kde = True)
```

```python
#Standardization using sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
#Covert to DataFrame
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[['total_bill']]),columns = ['total_bill'])
```

```python
#taking two features
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns = ['total_bill','tip'])
```

```python
#Normalization - Min Max Scaler
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
#scaling for new dataset
min_max.transform([[1.60,7.0,2.15]])
```

```python
#Unit vector
from sklearn.preprocessing import normalize
```

```python
pd.DataFrame(normalize(df[['distance','fare','tip']]))
```
